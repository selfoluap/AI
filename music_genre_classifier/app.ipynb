```python
from fastai.vision.all import *

# Define the path to the images_original folder
path = Path('/home/mrrobot/Projects/AI/music_genre_classifier/images_original')

# Define the data loader
dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, seed=42, item_tfms=Resize(224))

# Define the model architecture
learn = cnn_learner(dls, resnet34, metrics=accuracy)

# Train the model
learn.fine_tune(4)
```

```python
dls.show_batch()
```

```python
learn.show_results()
```

```python
interp = Interpretation.from_learner(learn)
interp.plot_top_losses(9, figsize=(15,10))
```

```python
learn.export("music_genre_classifier.pkl")
```
